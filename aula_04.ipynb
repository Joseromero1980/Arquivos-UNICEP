1.  **Entenda o problema:** Leia com atenção o que está sendo pedido.
2.  **Identifique as Entradas e Saídas:** O que eu recebo do usuário? O que eu devo mostrar na tela?

## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

## Exemplo Resolvido 2: Calculadora Simples

**O problema:** Peça dois números e um operador matemático (`+`, `-`, `*` ou `/`) ao usuário. Imprima o resultado da operação.

```python
num1 = float(input("Digite o primeiro número: "))
op = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

if op == "+":
    print(f"Resultado: {num1 + num2}")
elif op == "-":
    print(f"Resultado: {num1 - num2}")
elif op == "*":
    print(f"Resultado: {num1 * num2}")
elif op == "/":
    # Cuidados extras em programação: não se divide por zero!
    if num2 == 0:
        print("Erro: Divisão por zero.")
    else:
        print(f"Resultado: {num1 / num2}")
else:
    print("Operador inválido!")
```

### Desafio 3: Triângulos
Peça os 3 lados de um triângulo. O programa deverá informar se os valores formam um triângulo. Caso formem, diga o tipo do triângulo:
- Equilátero: 3 lados iguais;
- Isósceles: 2 lados iguais;
- Escaleno: 3 lados diferentes.
*Dica: Para formar um triângulo, a soma de quaisquer dois lados precisa ser maior que o terceiro lado.*

```python
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

if a + b > c and a + c > b and b + c > a:
    if a == b and b == c:
        print("Triângulo Equilátero")
    elif a == b or a == c or b == c:
        print("Triângulo Isósceles")
    else:
        print("Triângulo Escaleno")

```

### Exercício 8 (C): Validação de Data

Crie um programa em C que peça o dia, o mês e o ano. O programa deve verificar se a data é válida. Considere:
* Mês entre 1 e 12.
* Dia entre 1 e 31 (ou 30, 29, 28 dependendo do mês e se é bissexto).
* Anos bissextos (fevereiro tem 29 dias).

Use uma combinação de `if-else if-else` e operadores lógicos para validar todas as condições. Retorne "Data Válida" ou "Data Inválida".

```python
dia = int(input("Dia: "))
mes = int(input("Mês: "))
ano = int(input("Ano: "))

# verifica ano bissexto
bissexto = (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)

# valida mês
if mes < 1 or mes > 12:
    valido = False

# define limite de dias por mês
elif mes in [1, 3, 5, 7, 8, 10, 12]:
    valido = (dia >= 1 and dia <= 31)

elif mes in [4, 6, 9, 11]:
    valido = (dia >= 1 and dia <= 30)

elif mes == 2:
    if bissexto:
        valido = (dia >= 1 and dia <= 29)
    else:
        valido = (dia >= 1 and dia <= 28)

else:
    valido = False

if valido:
    print("Data Válida")
else:
    print("Data Inválida")
```

### Exercício 9 (C): Sistema de Acesso Multicamadas

Desenvolva um programa em C que simule um sistema de acesso. Peça ao usuário seu nome de usuário e um nível de permissão (um número inteiro de 1 a 3, onde 1 é 'básico', 2 é 'editor', 3 é 'administrador').

Defina as seguintes regras de acesso:
* Usuários com nível 3 (administrador) podem acessar todas as funcionalidades.
* Usuários com nível 2 (editor) podem acessar funcionalidades de edição e básicas, mas não as de administrador.
* Usuários com nível 1 (básico) podem acessar apenas funcionalidades básicas.
* Se o usuário for 'guest' (convidado), independentemente do nível de permissão, ele só pode acessar funcionalidades básicas.

Use `if-else if-else` e operadores lógicos para determinar qual conjunto de funcionalidades o usuário pode acessar e imprima a permissão de acesso (ex: "Acesso total (Administrador)", "Acesso de Edição e Básico", "Acesso Básico").

```python
usuario = input("Nome de usuário: ")
nivel = int(input("Nível de permissão (1 a 3): "))

if usuario == "guest":
    print("Acesso Básico")

elif nivel == 3:
    print("Acesso total (Administrador)")

elif nivel == 2:
    print("Acesso de Edição e Básico")

else:
    print("Acesso Básico")
```

### Exercício 10 (C): Classificador de Números Avançado

Escreva um programa em C que receba um número inteiro. Classifique-o da seguinte forma, utilizando uma única estrutura `if-else if-else` com expressões booleanas complexas:
* Se o número for positivo, par **E** maior que 100: "Muito Grande e Positivo Par"
* Se o número for positivo, ímpar **E** menor que 50: "Pequeno e Positivo Ímpar"
* Se o número for negativo **E** divisível por 3: "Negativo Múltiplo de Três"
* Se o número for zero: "É Zero"
* Caso contrário: "Outro Tipo de Número"

Certifique-se de que as condições não se sobreponham acidentalmente e que a ordem das verificações faça sentido lógico para cobrir todos os casos possíveis.

```python
num = int(input("Digite um número inteiro: "))

if num == 0:
    print("É Zero")

elif num > 0 and num % 2 == 0 and num > 100:
    print("Muito Grande e Positivo Par")

elif num > 0 and num % 2 != 0 and num < 50:
    print("Pequeno e Positivo Ímpar")

elif num < 0 and num % 3 == 0:
    print("Negativo Múltiplo de Três")

else:
    print("Outro Tipo de Número")
```
