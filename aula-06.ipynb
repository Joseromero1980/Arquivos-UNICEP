## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))

    if valor == 0:
        break # Sai do loop quando o usuário digitar zero

    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 1: Somatória dos Pares
Peça um número `N` ao usuário. Use um loop `for` ou `while` para calcular e exibir a SOMA de todos os números PARES entre 1 e `N`.

```python
# Seu código aqui

n = int(input("Digite um número: "))

soma = 0

for i in range(1, n + 1):
    if i % 2 == 0:  # verifica se é par
        soma += i

print(f"A soma dos números pares entre 1 e {n} é: {soma}")
```

```python
# Solução com while loop

n = int(input("Digite um número (para a solução com while loop): "))

soma_while = 0
contador = 1

while contador <= n:
    if contador % 2 == 0:
        soma_while += contador
    contador += 1

print(f"A soma dos números pares entre 1 e {n} (usando while) é: {soma_while}")
```

### Desafio 3.
Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
# Escreva seu código aqui

pop_a = 80000
pop_b = 200000

taxa_a = 0.03
taxa_b = 0.015

anos = 0

while pop_a < pop_b:
    pop_a += pop_a * taxa_a
    pop_b += pop_b * taxa_b
    anos += 1

print(f"Serão necessários {anos} anos para a população A ultrapassar ou igualar a população B.")
```

### Desafio 4.
Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

```python
# Escreva seu código aqui
while True:
    print("\n=== Simulação de Crescimento Populacional ===")

    pop_a = int(input("Digite a população do país A: "))
    pop_b = int(input("Digite a população do país B: "))

    taxa_a = float(input("Digite a taxa de crescimento do país A (ex: 0.03 para 3%): "))
    taxa_b = float(input("Digite a taxa de crescimento do país B (ex: 0.015 para 1.5%): "))

    # Validação básica
    if pop_a <= 0 or pop_b <= 0 or taxa_a <= 0 or taxa_b <= 0:
        print("Erro: valores devem ser positivos. Tente novamente.")
        continue

    anos = 0

    # Simulação
    while pop_a < pop_b:
        pop_a += pop_a * taxa_a
        pop_b += pop_b * taxa_b
        anos += 1

    print(f"\nSerão necessários {anos} anos para A alcançar ou ultrapassar B.")

    # Pergunta se deseja repetir
    repetir = input("\nDeseja repetir o programa? (s/n): ").lower()

    if repetir != "s":
        print("Programa encerrado.")
        break

```

```python
# Escreva seu código aqui

num = int(input("Digite um número inteiro: "))

fatorial = 1

print(f"{num}! = ", end="")

for i in range(num, 0, -1):
    fatorial *= i
    if i > 1:
        print(f"{i}.", end="")
    else:
        print(f"{i} = ", end="")

print(fatorial)
```

### Desafio 20.
Altere o programa de cálculo do fatorial, permitindo ao usuário calcular o fatorial várias vezes e limitando o fatorial a números inteiros positivos e menores que 16.

```python
# Escreva seu código aqui

while True:
    num = int(input("Digite um número inteiro positivo menor que 16 (ou -1 para sair): "))

    if num == -1:
        print("Encerrando o programa.")
        break

    if num < 0 or num >= 16:
        print("Valor inválido. Digite um número entre 0 e 15.")
        continue

    fatorial = 1
    print(f"{num}! = ", end="")

    for i in range(num, 0, -1):
        fatorial *= i
        if i > 1:
            print(f"{i}.", end="")
        else:
            print(f"{i} = ", end="")

    print(fatorial)
    print()

```

### Desafio 21.
Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1.

```python
# Escreva seu código aqui

num = int(input("Digite um número inteiro: "))

if num <= 1:
    print("Não é um número primo.")
else:
    primo = True

    for i in range(2, num):
        if num % i == 0:
            primo = False
            break

    if primo:
        print("É um número primo.")
    else:
        print("Não é um número primo.")
```

### Desafio 22.
Altere o programa de cálculo dos números primos, informando, caso o número não seja primo, por quais número ele é divisível.

```python
# Escreva seu código aqui

num = int(input("Digite um número inteiro: "))

if num <= 1:
    print("Não é um número primo.")
else:
    divisores = []

    for i in range(2, num):
        if num % i == 0:
            divisores.append(i)

    if len(divisores) == 0:
        print("É um número primo.")
    else:
        print("Não é um número primo.")
        print("É divisível por:", *divisores)

```

### Desafio 23.
Faça um programa que mostre todos os primos entre 1 e N sendo N um número inteiro fornecido pelo usuário. O programa deverá mostrar também o número de divisões que ele executou para encontrar os números primos. Serão avaliados o funcionamento, o estilo e o número de testes (divisões) executados.

```python
# Escreva seu código aqui

n = int(input("Digite um número inteiro N: "))

divisoes = 0

print("Números primos entre 1 e", n, ":")

for num in range(2, n + 1):
    primo = True

    for i in range(2, num):
        divisoes += 1  # conta cada teste de divisão

        if num % i == 0:
            primo = False
            break

    if primo:
        print(num, end=" ")

print("\n\nNúmero total de divisões executadas:", divisoes)

```

### Desafio 25.
Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da turma varia entre 0 e 25,26 e 60 e maior que 60; e então, dizer se a turma é jovem, adulta ou idosa, conforme a média calculada.

```python
# Escreva seu código aqui

n = int(input("Digite a quantidade de pessoas: "))

soma = 0

for i in range(1, n + 1):
    idade = int(input(f"Digite a idade da {i}ª pessoa: "))
    soma += idade

media = soma / n

print("\nMédia de idade da turma:", media)

if media >= 0 and media <= 25:
    print("Turma jovem")
elif media <= 60:
    print("Turma adulta")
else:
    print("Turma idosa")

```

### Desafio 26.
Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
# Escreva seu código aqui

votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0

n = int(input("Digite o número total de eleitores: "))

for i in range(1, n + 1):
    print(f"\nEleitor {i}, vote em um candidato:")
    print("1 - Candidato 1")
    print("2 - Candidato 2")
    print("3 - Candidato 3")

    voto = int(input("Seu voto: "))

    if voto == 1:
        votos_candidato1 += 1
    elif voto == 2:
        votos_candidato2 += 1
    elif voto == 3:
        votos_candidato3 += 1
    else:
        print("Voto inválido (não contabilizado).")

print("\nResultado da eleição:")
print("Candidato 1:", votos_candidato1, "votos")
print("Candidato 2:", votos_candidato2, "votos")
print("Candidato 3:", votos_candidato3, "votos")

```

### Desafio 27.
Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade de turmas e a quantidade de alunos para cada turma. As turmas não podem ter mais de 40 alunos.

```python
# Escreva seu código aqui

turmas = int(input("Digite a quantidade de turmas: "))

soma_alunos = 0

for i in range(1, turmas + 1):
    alunos = int(input(f"Digite a quantidade de alunos da turma {i} (máx. 40): "))

    while alunos < 0 or alunos > 40:
        alunos = int(input("Valor inválido. Digite um número entre 0 e 40: "))

    soma_alunos += alunos

media = soma_alunos / turmas

print("\nNúmero médio de alunos por turma:", media)
```
