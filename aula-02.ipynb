# Aula 02: Estruturas de Decisão

Até agora, nossos programas executaram todas as linhas de código, de cima para baixo. Mas e se a gente só quisesse executar um código se uma **condição** for verdadeira?
Para isso usamos as estruturas de decisão, lideradas pelo comando `if` (se).

## 1. O comando `if`

O `if` verifica se uma afirmação é Verdadeira (`True`) ou Falsa (`False`). Se for Verdadeira, o bloco de código dentro do `if` é executado.

```python
idade = 18

# Lê-se: Se idade for maior ou igual a 18...
if idade >= 18:
    print("Você é maior de idade.")
    print("Pode entrar na festa!")

print("Fim do programa.")
```

```python
idade = 45
# Lê-se: Se idade for maior ou igual a 45...
if idade >= 45:
  print("Você se aposentará em 20 anos")
  print("Tenha bom ânimo, você ainda poderá se desenvolver muito nesta empresa")

  print("Fim do programa")
```

```python
idade = 37

if idade >= 37:
    print("Parabéns pelo aniversário de empresa e obrigado por constribuir em todos esses anos. ")
    print("Receba este singelo presente em nome do grupo, como simbolo de gratidão, por toda sua jornada nesta empresa. Parabéns ")
print("fim do programa")
```

**Atenção à indentação!** No Brasil, chamamos de "espaçamento". O Python usa espaços no começo da linha (indentação) para saber quais linhas de código pertencem ao `if`. Tudo que estiver alinhado com espaços para dentro faz parte daquele bloco.

### Operadores de Comparação:
* `==` : Igual
* `!=` : Diferente
* `>`  : Maior que
* `<`  : Menor que
* `>=` : Maior ou igual a
* `<=` : Menor ou igual a

## 2. O comando `else` (Senão)

Se a condição do `if` não for verdadeira, podemos usar o `else` para realizar outra ação.

```python
nota = 5

if nota >= 7:
    print("Você foi Aprovado!")
else:
    print("Você foi Reprovado.")
```

```python
nota= 70

if nota >= 70:
    print("aprovado")
else:
    print("reprovado")
```

## 3. O comando `elif` (Senão Se)

Às vezes, temos mais de duas situações possíveis. Por exemplo, num semáforo de trânsito, temos verde, amarelo e vermelho. Usamos o `elif` para testar várias condições em sequência.

```python
cor = input("Digite a cor do semáforo (verde/amarelo/vermelho): ").lower()

if cor == "verde":
    print("Siga em frente!")
elif cor == "amarelo":
    print("Atenção, diminua a velocidade.")
elif cor == "vermelho":
    print("Pare!")
else:
    print("Cor inválida.")
```

```python
cor_temperatura = input("descubra a temperatura pela cor (azul/amarelo/vermelho): ").lower()

if cor_temperatura == "azul":
  print("atenção frio, se agasalhe!")
elif cor_temperatura == "amarelo":
  print("atenção: calor com possíbilidade de chuva, e queda na temperatura. Não esqueça seu guarda-chuva e blusa!")
elif cor_temperatura == "vermelho":
  print("hoje a temperatura varia de 28 a 33 graus. Hidrate-se bastante e use protetor sola!")
```

No código acima, usamos `.lower()` ao final do `input()` para converter o que o usuário digitou para letras minúsculas (assim identificamos 'Verde', 'VERDE' ou 'verde' da mesma forma).
