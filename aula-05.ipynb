# Aula 05: Estruturas de Repetição

Imagine que você precise escrever "Olá, Mundo!" 100 vezes. Você poderia escrever 100 `print` seguidos, mas isso daria muito trabalho. É para automatizar essas tarefas repetitivas que usamos as **Estruturas de Repetição**, ou **Laços (Loops)**.

## 1. O Laço `while` (Enquanto)

O `while` repete um bloco de código **ENQUANTO** uma condição for verdadeira. É como um `if` que, quando termina, volta para cima e checa a condição de novo.

```python
contador = 1

print("Vai começar a contagem!")
while contador < 5:
    # Esse bloco vai se repetir
    print(f"Contagem: {contador}")
    contador += 1 # MUITO IMPORTANTE! Atualizar a variável da condição

print("Terminou a contagem!")
```

### Looping Infinito
Se você esquecer de atualizar a variável dentro do `while` (como a linha `contador = contador + 1`), a condição sempre será Verdadeira, e seu programa nunca vai parar! Se isso acontecer no Jupyter, clique no botão "Stop" na barra superior ou Reinicie o Kernel.

## 2. O Laço `for` (Para)

O Python possui outro tipo de loop chamado `for`. Ele é extremamente flexível e serve para percorrer (iterar) coleções de itens (como listas de números, letras de uma palavra, ou arquivos em uma pasta).

### Usando o `for` com a função `range()`
A função `range()` gera uma sequência de números inteiros para trabalharmos. O `for` vai repetindo a partir do começo da sequência até o fim dela.

```python
# range(5) gera os números 0, 1, 2, 3 e 4.
# O Python sempre começa a contar do 0 e para um número antes do especificado.

for numero in range(5):
    print(f"O for passou no número {numero}")
```

```python
# Podemos dizer onde começar e onde terminar usando range(inicio, fim_não_incluso)
for i in range(1, 6):
    print(f"Passo {i} de 5")

print("---")

# Podemos também dizer os pulos usando range(inicio, fim, passo)
for i in range(2, 11, 2):
    print(f"Imprimindo números pares de 2 em 2: {i}")
```

### Analisando Textos Letra por Letra
Um texto em Python é tratado como uma coleção de letras. O comando `for` pode inspecionar letra por letra.

```python
palavra = "PYTHON"

for letra in palavra:
    print(f"Letra: {letra}")
```

## 3. Controlando as repetições: `break` e `continue`

Às vezes precisamos parar um loop no meio do caminho ou pular para o próximo item antes do código acabar.

*   `break`: Quebra (encerra) o loop imediatamente.
*   `continue`: Pula todo o código restante do loop atual, e vai direto para a próxima repetição.

```python
# Exemplo de break: O usuário digita senhas até acertar.
# Usamos o while True (que seria um loop infinito) e quebramos ele (break) quando der certo.

while True:
    senha = input("Qual é a senha secreta? (Dica: abracadabra):")

    if senha == "abracadabra":
        print("Acesso Liberado!")
        break # Sai do while

    print("Senha incorreta, tente de novo.")
```

```python
# Exemplo de continue: Imprimir de 1 a 6, mas pulando o 3 e o 4.

for i in range(1, 7):
    if i == 3 or i == 4:
        continue # Ignora o que está embaixo e volta para a próxima rodada do for.
    print(i)
```

## Exercício Prático

Crie um programa que faça uma contagem regressiva para estourar fogos.
Ele deve começar do 10 e ir até 1. Ao chegar no final, imprima 'BUM!!!'.
Use o `for` com o `range(10, 0, -1)` ou `while`.

```python

# Seu código aqui
print("Contagem regressiva para os fogos:")

for i in range(10, 0, -1):
    print(i)

print("BUM!!!")

```

```python
print("Contagem regressiva para os fogos:")

contador = 10

while contador > 0:
    print(contador)
    contador -= 1

print("BUM!!!")
```
