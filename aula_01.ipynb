## 3. Operações Matemáticas Básicas

Python é uma excelente calculadora! Veja os operadores básicos:

*   `+` : Adição
*   `-` : Subtração
*   `*` : Multiplicação
*   `/` : Divisão (resultado com casas decimais)
*   `//`: Divisão Inteira (resultado sem casas decimais)
*   `%` : Resto da divisão (Módulo)
*   `**`: Exponenciação (Potência)

```python
a = 10
b = 3

print(f"Soma: {a + b}")
print(f"Subtração: {a - b}")
print(f"Multiplicação: {a * b}")
print(f"Divisão: {a / b}")
print(f"Divisão Inteira: {a // b}")
print(f"Resto da Divisão: {a % b}")
print(f"Potência (10 elevado a 3): {a ** b}")
```

```python
a = 16
b = 144

print(f"soma: {a + b}")
```

## 4. Juntando Tudo: Funções Matemáticas + Entrada/Saída

Vamos fazer um programa que pede dois números ao usuário e calcula a soma. Como dissemos antes, o `input()` retorna um texto (string). Precisamos usar a função `int()` (para números inteiros) ou `float()` (para números com vírgula) para converter o texto em número.

```python
# Recebendo os números como texto (input) e logo em seguida convertendo para número inteiro (int)
numero1 = int(input("Digite o primeiro número inteiro: "))
numero2 = int(input("Digite o segundo número inteiro: "))

soma = numero1 + numero2

print(f"A soma de {numero1} com {numero2} é igual a {soma}.")
```

### Funções Matemáticas Mais Usadas (Módulo `math` e padrões)

Além dos operadores básicos, Python possui funções prontas para ajudar. Algumas já estão embutidas (`abs()`, `round()`), enquanto outras precisam ser chamadas de uma biblioteca matemática chamada `math`.

```python
import math

numero_negativo = -15
numero_quebrado = 3.74159

print(f"Valor Absoluto (tira o sinal): abs(-15) = {abs(numero_negativo)}")
print(f"Arredondamento: round(3.74159, 2 casas) = {round(numero_quebrado, 2)}")
print(f"Arredondamento para Cima: math.ceil(3.74159) = {math.ceil(numero_quebrado)}")
print(f"Arredondamento para Baixo: math.floor(3.74159) = {math.floor(numero_quebrado)}")
print(f"Raiz Quadrada de 9: math.sqrt(9) = {math.sqrt(9)}")
```

## 5. Gerador de Números Aleatórios (Sorteios)

Muitas vezes, em jogos ou sorteios, precisamos que o computador escolha um número aleatório (sortido) por nós. O Python tem uma biblioteca interna chamada `random` exatamente para isso.

```python
import random

# Sorteando um número inteiro entre 1 e 10 (incluindo o 10)
numero_sorteado = random.randint(1, 10)
print(f"O número sorteado de 1 a 10 foi: {numero_sorteado}")

# Escolhendo um elemento aleatório de uma lista de opções (veremos listas mais a frente, mas é útil saber!)
opcoes = ["Cara", "Coroa"]
moeda = random.choice(opcoes)
print(f"Joguei a moeda e deu: {moeda}")
```

```python
numero_sorteado = random.randint(1, 50000)
print(f"o numero sorteado de 1 a 50000 foi: {numero_sorteado}" )

opcoes = ["bolão", "sozinho"]
moeda = random.choice(opcoes)
print(f"joguei na loteria e ganhei: {moeda}")
```

## Exercício Prático

Agora é sua vez! Crie uma pequena calculadora de Índice de Massa Corporal (IMC).
1. Peça ao usuário seu peso em KG (ex: 75.5) -> Lembre-se, use `float()`
2. Peça ao usuário sua altura em Metros (ex: 1.80) -> Use `float()`
3. Calcule o IMC (Fórmula: peso dividido pela altura ao quadrado)
4. Imprima o resultado na tela usando `f-string`, formatado com 2 casas decimais.

Use a célula de código abaixo para resolver:

```python
# Escreva o código do seu exercício aqui!

# Solicita os dados ao usuário
peso = float(input("Digite seu peso em kg (ex: 75.5): "))
altura = float(input("Digite sua altura em metros (ex: 1.80): "))

# Calcula o IMC
imc = peso / (altura ** 2)

# Exibe o resultado formatado
print(f"Seu IMC é: {imc:.2f}")

```
